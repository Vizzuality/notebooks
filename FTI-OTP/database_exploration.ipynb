# Observation document extraction

```python
from IPython.display import display, Markdown
import pandas as pd
import binascii
import os
import gc
```

```python
path_base = os.getcwd().split('/work/')[0]
data_path = os.path.join(path_base,'work','data','fti','dst') 
exit_path = os.path.join(data_path,'documents')

def document_extractor(dataset):
    data = pd.concat((chunk for chunk in pd.read_csv(dataset['path'], chunksize=50, low_memory  = True, iterator= True, sep=';', quotechar="'")),ignore_index=True) 
    section= os.path.join(exit_path,dataset['file_name'].split('.')[0]) 
    try:
        display(Markdown('## '+dataset['file_name'].split('.')[0]))
        os.makedirs(section)
        display(data.head(2))
        data_columns= [{"name": column,"data":column.rstrip('FileName')} for column in data.columns.get_values() if 'DocumentFileName'in column or 'PhotoFileName' in column]
        for c in data_columns:
            for d, row in data.iterrows():
                path = section +'/' + str(row[c['name']])
                if type(row[c['data']]) is str:
                    with open(path, "wb") as f:
                        f.write(binascii.unhexlify(row[c['data']].encode()[2:]))
        print('done')
        del data
    except OSError:
        if not os.path.isdir(path):
            raise
```

```python
my_files=[]
for root, dirs, files in os.walk(data_path):
    for file in files:
        if file.endswith('csv'):
            my_files.append({"file_name":file, "path":os.path.join(root, file)})   
```

```python
for file in my_files:
    document_extractor(file)
```
