```python
import pandas as pd
import binascii
```

```python
url = 'ObservationReports.csv'
data = pd.read_csv(url,delimiter=';')
```

```python
data.Document.count()
```

```python
#extract data as pdf or wathver file stored
for d, row in data.iterrows():
    print(d)
    path = 'dst/' + str(row['DocumentFileName'])
    if type(row['Document'])is str:
        with open(path, "wb") as f:
            f.write(binascii.unhexlify(row['Document'].encode()[2:]))
```
