```python
a = pd.read_csv("./final/uganda2013/version_3/completo.csv", sep=',', low_memory=False, index_col='ID')
```

```python
a['age_group'] = ''
a.loc[(a['PN8'] < 18), ['age_group']] = '<18'
a.loc[(a['PN8'] >= 18) & (a['PN8'] < 25 ), ['age_group']] = '18-25'
a.loc[(a['PN8'] >= 25) & (a['PN8'] < 35 ), ['age_group']] = '25-35'
a.loc[(a['PN8'] >= 35) & (a['PN8'] < 45 ), ['age_group']] = '35-45'
a.loc[(a['PN8'] >= 45) & (a['PN8'] < 60 ), ['age_group']] = '45-60'
a.loc[(a['PN8'] >= 60), ['age_group']] = '>60'
a.loc[(a['PN8'] == np.nan), ['age_group']] = 'Not respond'
```

```python
a.loc[(a['FAS'] != 'Excluded'), ['FAS']] = ''
a.loc[(a['Saving_Strand'] != 'Not saving'), ['Saving_Strand']] = ''
a.loc[(a['Remittances_Strand'] != 'Not remitting'), ['Remittances_Strand']] = ''
a.loc[(a['Insurance_Strand'] != 'Not insured'), ['Insurance_Strand']] = ''
a.loc[(a['Credit_Strand'] != 'Not borrowing'), ['Credit_Strand']] = ''
```

```python
a.head(1)
```

```python
a.to_csv("./final/uganda2013/version_3/completo.csv", sep=',')
```
